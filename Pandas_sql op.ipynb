```python
# 1) Missing Data
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'CL1' : [170, 50, np.nan],
                  'CL2' : [168, np.nan , np.nan],
                  'CL3' : [180, 80, 91]})
df
```

```python
df.dropna(thresh=2)
```

```python
df.fillna(value = 0)
df
```

```python
df['CL1'].fillna(value = df['CL1'].mean())
```

```python
# 2) Panda Groupby
```

```python
data = {'company' : ['GOUG' , 'GOUG' , 'MSFT' , 'MSFT' , 'FB' , 'FB'],
        'Person'  : ['Sam', 'Charlie', 'Amy', 'Vanessa' , 'Carl' , 'Sarah'],
        'Sales' : [200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.groupby('company')['Sales'].min()
#.loc['FB']

```

```python
df.groupby('company')['Sales'].describe().transpose()
```

```python
# 3) Panda-Merging, Joining, and Concatenating
```

```python
import pandas as pd
import numpy as pd
```

```python
df1 = pd.DataFrame({ 'A' : ['A0', 'A1', 'A2', 'A3'],
                     'B' : ['B0', 'B1', 'B2', 'B3'],
                     'C' : ['C0', 'C1', 'C2', 'C3'],
                     'D' : ['D0', 'D1', 'D2', 'D3']
},
index = [0,1,2,3])

df2 = pd.DataFrame({ 'A' : ['A4', 'A5', 'A6', 'A7'],
                     'B' : ['B4', 'B5', 'B6', 'B7'],
                     'C' : ['C4', 'C5', 'C6', 'C7'],
                     'D' : ['D4', 'D5', 'D6', 'D7']
},
index = [4,5,6,7])

df3 = pd.DataFrame({ 'A' : ['A8', 'A9', 'A10', 'A11'],
                     'B' : ['B8', 'B8', 'B9', 'B10'],
                     'C' : ['C8', 'C9', 'C10', 'C11'],
                     'D' : ['D8', 'D9', 'D10', 'D11']
},
index = [8,9,10,11])
```

```python
df1
```

```python
pd.concat([df1,df2,df3])
```
